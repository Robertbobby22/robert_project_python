LOADING THE DATASET

```python
from datasets import load_dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df
```

UPDATING job_posted_date FROM Object to Datetime

```python
import pandas as pd
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
df.info()
```

PLOTING Minimum SALARIES PAID Yearly

```python
import matplotlib.pyplot as plt
job_titles = df.groupby('job_title_short')['salary_year_avg'].min()
job_titles = job_titles.sort_values(ascending=True)
plt.title('THE MINIMUM SALARIES OF DIFFERENT JOB TITLES Yearly')
plt.ylabel('Job Titles')
plt.xlabel('Minimum Salaries Yearly (usd)')
plt.barh(job_titles.index, job_titles)
```

PLOTING MEDIAN SALARIES PAID Yearly

```python
job_titles = df.groupby('job_title_short')['salary_year_avg'].median()
job_titles = job_titles.sort_values()
plt.title('THE MEDIAN SALARIES OF DIFFERENT JOB TITLES Yearly')
plt.ylabel('Job Titles')
plt.xlabel('Median Salaries Yearly (USD)')
plt.barh(job_titles.index, job_titles)
```

PLOTING MINIMUM SALARIES PAID Hourly

```python
job_titles = df.groupby('job_title_short')['salary_hour_avg'].min()
job_titles = job_titles.sort_values()
plt.title('JOB TITLES AND MINIMUM SALARIES PAID Hourly')
plt.ylabel('Job Titles')
plt.xlabel('Minimum Hourly Salaries (USD)')
plt.barh(job_titles.index, job_titles)
```

PLOTING MEDIAN SALARIES PAID Hourly

```python
job_titles = df.groupby('job_title_short')['salary_hour_avg'].median()
job_titles = job_titles.sort_values(ascending=True)
plt.title('JOB TITLES AND MEDIAN SALARIES PAID Hourly')
plt.ylabel('Job Titles')
plt.xlabel('Median Salaries Paid Hourly (USD)')
plt.barh(job_titles.index, job_titles)
```
